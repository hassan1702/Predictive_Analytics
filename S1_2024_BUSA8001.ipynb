### Forecasting Problem

The primary objective of this analysis is to forecast the prices of listed used cars based on various car features provided in the dataset. This task is a supervised learning problem, specifically a regression analysis, where the goal is to predict a continuous target variable - the car price. Accurate car price forecasting has significant implications for a wide range of stakeholders, including buyers, sellers, insurance companies, financial institutions, and car dealers importing cars.

**Buyers**

For buyers, not knowing the correct price of a used car can lead to overpaying or missing out on better deals. Accurate price forecasts enable buyers to compare prices across different listings and models, helping them make informed purchasing decisions and ensuring they get the best value for their money.

**Sellers**

Sellers face challenges in setting competitive prices without accurate price information. Incorrect pricing can either deter buyers if the price is too high or result in lost revenue if the price is too low. Accurate price forecasts help sellers optimize their pricing strategy, leading to quicker sales and better inventory management.

**Insurance Companies**

Insurance companies need accurate car valuations to calculate premiums and process claims. Inaccurate price forecasts can lead to premiums that do not reflect the car's true value, either overcharging or undercharging customers. Accurate predictions ensure fair premium calculations and compensation in the event of claims.

**Financial Institutions**

Financial institutions rely on accurate car valuations for loan approvals and asset management. Incorrect valuations can result in lending too much or too little, affecting risk management and customer satisfaction. Accurate price forecasts help banks and lenders assess the collateral value of vehicles, ensuring appropriate loan amounts.

**Car Dealers Importing Cars**

Car dealers importing cars from abroad need accurate price forecasts to set competitive local market prices. Understanding the market value of imported cars helps dealers determine if import costs can be covered while still making a profit. Accurate predictions enable dealers to choose the right models to import and set attractive prices for buyers.

### Evaluation Criteria

The performance of the forecasting models in this competition is assessed using the Mean Absolute Percentage Error (MAPE). MAPE is a popular metric for evaluating the accuracy of regression models, particularly in forecasting:

$$ \text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100 $$

where:
- \( y_i \) is the actual value,
- \( \hat{y}_i \) is the predicted value,
- \( n \) is the number of observations.

Other criteria to consider are the Root Mean Squared Error (RMSE) that measures the square root of the average of squared differences between predicted and actual values, emphasizing larger errors; and, R-squared (R²). This one Indicates the proportion of the variance in the dependent variable that is predictable from the independent variables, showing the model's explanatory power. These metrics will provide a comprehensive assessment of the model’s performance.

**Task 2, Question 2** Create at least 5 new features from the existing numerical variables which contain multiple items of information, for example you could extract maximum torque and torque rpm from the torque variable.  
(2 marks)

```python
def extract_torque_features(df):
    df['torque'] = df['torque'].astype(str).fillna('0 Nm @ 0')
    # Improved regex to handle variations in formatting
    df[['max_torque', 'torque_rpm']] = df['torque'].str.extract(r'(\d+\.?\d*)\s*lb-ft\s*@\s*(\d{1,3}(?:,\d{3})*|\d+)')
    df['max_torque'] = df['max_torque'].astype(float)
    df['torque_rpm'] = df['torque_rpm'].str.replace(',', '').astype(float)
    return df

# Apply the function to both train and test datasets
train_df = extract_torque_features(train_df)
test_df = extract_torque_features(test_df)

# Convert 'power' column to string type
train_df['power'] = train_df['power'].astype(str)
test_df['power'] = test_df['power'].astype(str)

# Extract the RPM part from the 'power' column and handle NaN values
train_df['power_rpm'] = train_df['power'].str.extract(r'@ (\d{1,3}(?:,\d{3})*|\d+)')[0].str.replace(',', '').fillna(0).astype(int)
test_df['power_rpm'] = test_df['power'].str.extract(r'@ (\d{1,3}(?:,\d{3})*|\d+)')[0].str.replace(',', '').fillna(0).astype(int)


# Create additional new features
current_year = datetime.now().year
train_df['car_age'] = current_year - train_df['year']
test_df['car_age'] = current_year - test_df['year']

# Create a new feature for mileage per year
train_df['mileage_per_year'] = train_df['mileage'] / train_df['car_age']
test_df['mileage_per_year'] = test_df['mileage'] / test_df['car_age']

# Ensure dimensions columns are numerical
train_df['length'] = train_df['length'].astype(float)
test_df['length'] = test_df['length'].astype(float)
train_df['width'] = train_df['width'].astype(float)
test_df['width'] = test_df['width'].astype(float)
train_df['height'] = train_df['height'].astype(float)
test_df['height'] = test_df['height'].astype(float)

# Create a new feature for horsepower to size (using length, width, and height as size proxies)
train_df['hp_to_size'] = train_df['horsepower'] / (train_df['length'] * train_df['width'] * train_df['height'])
test_df['hp_to_size'] = test_df['horsepower'] / (test_df['length'] * test_df['width'] * test_df['height'])

# Display the first few rows to verify new features
new_features = [
    'max_torque', 'torque_rpm', 'car_age',
    'mileage_per_year', 'hp_to_size', 'power_rpm'
]

print("New features in train data:")
print(train_df[new_features].head())

print("New features in test data:")
print(test_df[new_features].head())
```

We created 6 new features from the existing numerical variables in the train and test datasets. First, we extracted max_torque and torque_rpm from the torque column using regex to handle different formatting styles and converted these values to floats. Next, we converted the power column to string type and extracted the RPM part, filling any NaN values with 0, which we will impute later on to actual numbers, and converting the results to integers. We calculated the car's age by subtracting the manufacturing year from the current year. Additionally, we ensured the dimensions (length, width, height) were numeric and created a hp_to_size feature by dividing horsepower by the product of length, width, and height. 

**Task 2, Question 3**: Impute the missing values for all features in both the training and test datasets.   
(3 marks)

```python
# Identify numerical and categorical columns
numerical_cols = train_df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = train_df.select_dtypes(include=['object', 'bool']).columns

# Exclude 'price' from numerical columns for the test dataset
numerical_cols = numerical_cols.drop('price', errors='ignore')

# Function to impute missing values
def impute_missing_values(train, test, numerical_cols, categorical_cols):
    # Impute numerical columns
    num_imputer = SimpleImputer(strategy='mean')
    train[numerical_cols] = num_imputer.fit_transform(train[numerical_cols])
    test[numerical_cols] = num_imputer.transform(test[numerical_cols])
    
    # Impute categorical columns
    cat_imputer = SimpleImputer(strategy='most_frequent')
    train[categorical_cols] = cat_imputer.fit_transform(train[categorical_cols])
    test[categorical_cols] = cat_imputer.transform(test[categorical_cols])
    
    return train, test

# Apply imputation to both train and test datasets
train_df, test_df = impute_missing_values(train_df, test_df, numerical_cols, categorical_cols)

# Verify imputation by checking for any remaining missing values
print("Missing values in training data after imputation:")
print(train_df.isnull().sum().sum())

print("Missing values in test data after imputation:")
print(test_df.isnull().sum().sum())
```

In the imputation process, we first identified numerical and categorical columns, excluding the price column from numerical columns since it is the target variable. We used the SimpleImputer with the mean strategy for numerical columns and the most frequent strategy for categorical columns to handle missing values. After imputation, we verified that there were no remaining missing values in both the training and test datasets.

```python
# Define the categorical columns to encode
categorical_columns = [
    'body_type', 'franchise_dealer', 'fuel_type', 
    'is_new', 'listing_color', 'transmission', 
    'wheel_system',
]

# Define a function to keep only top 5 categories and replace the rest with 'other'
def keep_top_categories(df, cols):
    for col in cols:
        # Get the top 5 categories
        top5 = df[col].value_counts().index[:5]
        
        # Replace categories not in top 5 with 'other'
        df.loc[~df[col].isin(top5), col] = 'other'
    return df

# Apply the function to the train and test data
train_df = keep_top_categories(train_df, categorical_columns)
test_df = keep_top_categories(test_df, categorical_columns)

# Initialize OneHotEncoder with handle_unknown set to 'ignore'
encoder = OneHotEncoder(sparse=False, drop='first', handle_unknown='ignore')

# Fit and transform the train data
encoded_train = encoder.fit_transform(train_df[categorical_columns])
encoded_test = encoder.transform(test_df[categorical_columns])

# Create DataFrame from encoded data
encoded_train_df = pd.DataFrame(encoded_train, columns=encoder.get_feature_names_out(categorical_columns))
encoded_test_df = pd.DataFrame(encoded_test, columns=encoder.get_feature_names_out(categorical_columns))

# Concatenate the encoded columns back to the original dataframe
train_df = pd.concat([train_df.drop(columns=categorical_columns), encoded_train_df], axis=1)
test_df = pd.concat([test_df.drop(columns=categorical_columns), encoded_test_df], axis=1)
```

```python
# Initialize the model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Cross-validation score
cv_scores_rf = cross_val_score(rf, selected_features_train, target_train, cv=5, scoring='r2')
print(f"Random Forest Cross-Validation R² Score: {cv_scores_rf.mean()}")

# Train the model
rf.fit(selected_features_train, target_train)

# Predict on the test set
price_predictions_rf = rf.predict(selected_features_test)

# Evaluate the model on the training set
train_r2_rf = r2_score(target_train, rf.predict(selected_features_train))
print(f"Random Forest Training R² Score: {train_r2_rf}")
```

The Random Forest Regressor model was initialized and evaluated for predicting used car prices. After training the model on the full training dataset, the training R² score was found to be 0.9796, showcasing an excellent fit on the training data. These results suggest that the Random Forest model is highly effective in capturing the variability in used car prices, with a good balance between training performance and cross-validation scores.

```python
# Initialize the model
gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)

# Cross-validation score
cv_scores_gb = cross_val_score(gb, selected_features_train, target_train, cv=5, scoring='r2')
print(f"Gradient Boosting Cross-Validation R² Score: {cv_scores_gb.mean()}")

# Train the model
gb.fit(selected_features_train, target_train)

# Predict on the test set
price_predictions_gb = gb.predict(selected_features_test)

# Evaluate the model on the training set
train_r2_gb = r2_score(target_train, gb.predict(selected_features_train))
print(f"Gradient Boosting Training R² Score: {train_r2_gb}")
```

The Gradient Boosting Regressor model was initialized and evaluated for predicting used car prices.
